**To get the current working directory and assigning to the project_path**

```python
project_path = Path.cwd().parent
project_path
```

 **.csv is a standard way to store structured data in a plain text format.
     And it is assigned and stored in data_path** 


```python
data_path = project_path/"data"/"weatherHistory.csv"
data_path
```

**read_csv is used to read the content in the data path and it is stored in the data frame.
df.head(2) is used to get the information of first two rows.**


```python
df = pd.read_csv(data_path)
df.head(2)
```

**df.tail(2) to get the information of bottom two rows.**

```python
df.tail(2)
```

**Changing the data type of the formatted data from string to datetime**

```python
#Changing Formatted Date from String to Datetime
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df['Formatted Date'][0]
```

```python
df = df.sort_values(by = 'Formatted Date')
df.head(2)
```

**info() is used to get the details of all the rows and columns along with its respective data type.**

```python
df.info()
```

**describe() method returns description of the data in the DataFrame.**

```python
df.describe()
```

**The columns which are inputted in to our model is called as features.
We can select multiple features by providing a (list of columns names) inside the bracket.
For example In the below code line it gives the information about the formatted data,Apparent temperature,humidity.
th.head() by default it takes the value as 5.**


```python
features = ["Formatted Date", "Apparent Temperature (C)", "Humidity"]
th = df[features]
th.head()
```

**set_index() is a method to set a List, Series or Data frame as index of a Data Frame.
Index column can be set while making a data frame too. 
But sometimes a data frame is made out of two or more data frames and hence later index can be changed using this method.**

**Now we can see here, the formatted Data is acting as the index.**

```python
th = th.set_index('Formatted Date')
th.head()
```

**For our problem statement we need to delete the 2005 row details.
Then the th.head() is used to see the changes.**

```python
# delete 2005 row 
th = th.iloc[1:]
th.head()
```

**Now we can see that we are having the information of two columns Apparent Temperature (C),Humidity  with the datatype float.**

```python
th.info()
```

**describe() method returns description of the data in the DataFrame.**

```python
th.describe()
```

**Plotting the line graph for the given data**

```python
sns.lineplot(data=th);
```

**Resampling (M) and we are taking mean in the form of monthly status**

```python
montly_th = th.resample('M').mean()
montly_th.head()
```

**To get the detailed information about the data**

```python
montly_th.info()
```

**Variation of the temperature and humididty over the period of time**

```python
#define plotting region (2 rows, 2 columns)
fig, axes = plt.subplots(1, 2, figsize= (20, 7))

#create boxplot in each subplot
sns.lineplot(data=montly_th, x='Formatted Date', y='Apparent Temperature (C)', ax=axes[0])
sns.lineplot(data=montly_th, x='Formatted Date', y='Humidity', ax=axes[1])
plt.show()
```

**We need to take the samples of each month corresponds to every year**

```python
#retrieving the data of a particular month from every year
month_idx = 4 # April
montly_th_april = montly_th[montly_th.index.month == month_idx]
montly_th_april
```

**graph of temperature and humidity of the particular month**

```python
plt.figure(figsize=(15, 5))

plt.plot(montly_th_april, marker='o', label=['Apparent Temperature (C)','Humidity'] );
plt.legend(loc = 'center right', fontsize = 10)
plt.title('Relation between temperature and humidity for the month of April')
plt.show()
```

**By the above graph we can say there is a drastic increase in temperature wrt to humididty at the year 2009.**

## Conclusion ##
